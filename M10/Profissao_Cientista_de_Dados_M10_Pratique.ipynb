```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# Função para converter os valores de '+10mil' para números
def converter_vendas(valor):
    # Remove o símbolo de "+" e substitui "mil" por "000", além de transformar 'Nenhum' em 0
    # Isso é útil para lidar com valores de vendas em formatos não numéricos no CSV
    valor = valor.replace('+', '').replace('mil', '000').replace("Nenhum",'0')
    # Converte a string processada em um número inteiro
    return int(valor)
```

```python
# Lendo o arquivo CSV
df = pd.read_csv('ecommerce_preparados.csv')

# Exibindo as primeiras linhas do dataframe para inspecionar os dados
df.head()
```

```python
# Lendo o arquivo CSV e armazenando em um DataFrame
df = pd.read_csv('ecommerce_preparados.csv')

# Exibindo as primeiras linhas do DataFrame para inspecionar os dados
df.head()

# Verificando a existência de valores ausentes em cada coluna
print(df.isnull().sum())

# Exibindo estatísticas descritivas das colunas numéricas
# Isso permite entender a distribuição dos dados, incluindo média, mínimo e máximo
print(df.describe())

# Exibindo os tipos de dados de cada coluna
print(df.dtypes)

# Exibindo os nomes de todas as colunas do DataFrame
# Útil para garantir que estamos cientes de todas as variáveis disponíveis
print(df.columns)
```

# **Gráficos solicitados**

## **Gráfico de Dispersão** (Relação entre Preço e Nota)




```python
# Criando um gráfico de dispersão para visualizar a relação entre o preço e a nota do produto
# O gráfico de dispersão é útil para entender se há alguma correlação entre essas duas variáveis
plt.figure(figsize=(10,6))
sns.scatterplot(x='Preço', y='Nota', data=df)
plt.title('Dispersão entre Preço e Nota')
plt.xlabel('Preço')  # Rótulo do eixo X
plt.ylabel('Nota')   # Rótulo do eixo Y
plt.show()
```

## **Mapa de Calor** (Correlação entre Nota, Avaliações, Preço, Desconto e Vendas)



```python
# Selecionando as variáveis numéricas que são relevantes para análise de correlação
# Aqui escolhemos as colunas que podem ter uma relação entre si, como:
# Nota (avaliação do produto), número de avaliações, desconto, preço e quantidade vendida
variaveis_interessantes = ['Nota', 'N_Avaliações', 'Desconto', 'Preço', 'Qtd_Vendidos']

# Criando uma cópia explícita do DataFrame original
# Isso previne alterações não intencionais no DataFrame original, pois df_selecionado será uma cópia independente
df_selecionado = df[variaveis_interessantes].copy()

# Convertendo a coluna 'Qtd_Vendidos' para valores numéricos
# Aqui usamos a função 'converter_vendas' para transformar strings (como '+10mil') em números inteiros
# Usamos .loc[:, 'Qtd_Vendidos'] para garantir que a modificação ocorra sem o aviso de 'SettingWithCopyWarning'
df_selecionado.loc[:, 'Qtd_Vendidos'] = df_selecionado['Qtd_Vendidos'].apply(converter_vendas)

# Calculando a matriz de correlação entre as variáveis selecionadas
# A correlação quantifica a relação linear entre duas variáveis (varia de -1 a 1)
# Um valor de correlação próximo de 1 indica uma relação linear positiva, enquanto -1 indica uma relação linear negativa
correlacao = df_selecionado.corr()

# Gerando o mapa de calor para visualizar as correlações entre as variáveis
# O mapa de calor permite visualizar facilmente a intensidade das correlações, com uma escala de cores
# O argumento 'annot=True' insere os valores das correlações diretamente nas células
# Usamos o colormap 'coolwarm' para facilitar a distinção entre correlações positivas e negativas
# vmin=-1 e vmax=1 estabelecem a escala de cores variando de correlação mínima (-1) até a máxima (1)
plt.figure(figsize=(10,8))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Mapa de Calor: Correlação entre Nota, Avaliações, Preço, Desconto e Vendas')
plt.show()
```

## **Gráfico de Barras** (Participação das 10 Marcas com Mais Vendas)

```python
# Criando uma cópia do DataFrame original para modificações
# Isso é útil para manter o DataFrame original inalterado
df_vendas = df.copy()
df_vendas['Qtd_Vendidos'] = df_vendas['Qtd_Vendidos'].apply(converter_vendas)

# Agrupando por marca e somando as quantidades vendidas para cada uma
# Aqui estamos agregando os dados de vendas por marca
df_barras = df_vendas.groupby('Marca')['Qtd_Vendidos'].sum().reset_index()

# Ordenando para pegar as 10 marcas com mais vendas
# Estamos interessados apenas nas marcas com maiores volumes de venda
df_barras_top10 = df_barras.sort_values(by='Qtd_Vendidos', ascending=False).head(10)

# Criando o gráfico de barras para as 10 principais marcas
# Agora utilizamos a variável 'Marca' no argumento 'hue' e definimos legend=False
plt.figure(figsize=(10,6))
ax = sns.barplot(x='Marca', y='Qtd_Vendidos', data=df_barras_top10, hue='Marca', palette='viridis', legend=False)

# Adicionando os números de vendas no topo de cada barra
# Isso facilita a leitura dos valores de venda diretamente no gráfico
for i in ax.containers:
    ax.bar_label(i, label_type='edge', fmt='%d')

# Ajustando os rótulos e o título para melhor visualização
plt.title('Top 10 Marcas com Mais Produtos Vendidos')
plt.xlabel('Marca')  # Rótulo do eixo X
plt.ylabel('Quantidade Vendida')  # Rótulo do eixo Y
plt.xticks(rotation=45)  # Rotacionando os rótulos do eixo X para evitar sobreposição
plt.show()
```

## **Gráfico de Pizza** (Participação por Faixas de Preço)

```python
# Função para categorizar os produtos em faixas de preço
# Cada faixa representa uma gama específica de valores de preço
def faixa_preco(preco):
    if preco < 50:
        return 'abaixo de 50'
    elif 50 <= preco < 100:
        return '50 a 99'
    elif 100 <= preco < 150:
        return '100 a 149'
    elif 150 <= preco < 200:
        return '150 a 199'
    elif 200 <= preco <= 300:
        return '200 a 300'
    else:
        return 'acima de 300'

# Criando uma nova coluna 'Faixa_Preco' aplicando a função anterior
# Isso nos ajuda a agrupar os produtos em diferentes faixas de preço
df['Faixa_Preco'] = df['Preço'].apply(faixa_preco)

# Contando quantos produtos estão em cada faixa de preço
# Essa contagem será usada para gerar o gráfico de pizza
df_pizza_precos = df['Faixa_Preco'].value_counts().reset_index()
df_pizza_precos.columns = ['Faixa_Preco', 'Quantidade']

# Criando o gráfico de pizza para mostrar a participação de cada faixa
# O gráfico de pizza é útil para visualizar a proporção de cada categoria
plt.figure(figsize=(10, 6))
plt.pie(df_pizza_precos['Quantidade'], labels=df_pizza_precos['Faixa_Preco'],
        autopct=lambda p: f'{p:.1f}%' if p > 0 else '', startangle=90,
        colors=sns.color_palette('viridis', len(df_pizza_precos)))
plt.title('Participação de Produtos por Faixa de Preço')
plt.axis('equal')  # Para garantir que o gráfico seja um círculo perfeito
plt.show()
```

## **Gráfico de Densidade** (Distribuição do Preço)

```python
# Criando uma cópia do DataFrame original para modificações
# Isso nos permite trabalhar com os dados sem alterar o DataFrame original
df_vendas_densidade = df.copy()

# Aplicando a função de conversão para transformar 'Qtd_Vendidos' em números inteiros
df_vendas_densidade['Qtd_Vendidos'] = df_vendas_densidade['Qtd_Vendidos'].apply(converter_vendas)

# Criando o gráfico de densidade KDE para mostrar a densidade entre preço e quantidade vendida
# O gráfico de densidade nos permite visualizar onde estão concentradas as vendas em relação ao preço
plt.figure(figsize=(10, 6))
sns.kdeplot(x=df_vendas_densidade['Preço'], y=df_vendas_densidade['Qtd_Vendidos'], fill=True, cmap='coolwarm', thresh=0)
plt.title('Densidade de Vendas em Relação ao Preço')
plt.xlabel('Preço')  # Rótulo do eixo X
plt.ylabel('Quantidade Vendida')  # Rótulo do eixo Y

# Ajustando os limites dos eixos X e Y para dar uma margem aos dados
plt.ylim(df_vendas_densidade['Qtd_Vendidos'].min(), df_vendas_densidade['Qtd_Vendidos'].max() + 5000)
plt.xlim(df_vendas_densidade['Preço'].min(), df_vendas_densidade['Preço'].max() + 50)

plt.show()
```
