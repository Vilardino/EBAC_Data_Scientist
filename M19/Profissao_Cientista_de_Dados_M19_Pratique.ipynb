**Efetividade de Duas Estratégias de Ensino**

Imagine que uma escola esteja avaliando a eficácia de duas estratégias de ensino de matemática para alunos do ensino médio. Eles querem determinar se há uma diferença significativa no desempenho médio dos alunos entre as duas estratégias.

# **Hipóteses:**

* Hipótese nula (H0): A média das notas dos alunos na estratégia A é igual à média das notas dos alunos na estratégia B.
* Hipótese alternativa (H1): A média das notas na Estratégia B é maior do que a média das notas na Estratégia A.

# **Dados:**

* Amostra da Estratégia A: Notas de 50 alunos que receberam a Estratégia A.
* Amostra da Estratégia B: Notas de 50 alunos que receberam a Estratégia B.

Usaremos um teste Z para comparar as médias das notas entre as duas amostras.

Se o p-valor do teste Z for menor que um nível de significância pré-determinado (por exemplo, α = 0.05), rejeitamos a hipótese nula e concluímos que há uma diferença significativa nas médias das notas entre as duas estratégias de ensino.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

Os dados são criados a seguir:

```python
# Definindo médias e desvios padrão para as notas nas duas estratégias
media_estrategia_A = 70
desvio_padrao_estrategia_A = 10

media_estrategia_B = 75
desvio_padrao_estrategia_B = 12

# Gerando as amostras de notas para cada estratégia de ensino da nossa base
np.random.seed(0)  # Para garantir a reprodutibilidade dos resultados
amostra_estrategia_A = np.random.normal(loc=media_estrategia_A, scale=desvio_padrao_estrategia_A, size=50)
amostra_estrategia_B = np.random.normal(loc=media_estrategia_B, scale=desvio_padrao_estrategia_B, size=50)

print("Notas da Estratégia A:", amostra_estrategia_A[:5])
print("Notas da Estratégia B:", amostra_estrategia_B[:5])
```

**Justificativa:** A hipótese alternativa (H1) estabelece que a média das notas da Estratégia B é maior do que a média das notas da Estratégia A. Dessa forma, o foco está em valores que indiquem um aumento na média da Estratégia B em relação à Estratégia A, o que direciona a análise para o lado direito da distribuição.

Caso a análise envolvesse a verificação de diferenças em ambas as direções (maior ou menor), o teste seria bicaudal. Contudo, neste caso específico, a ênfase está no aumento, o que caracteriza o teste como unilateral à direita.

# 2) Calcule as médias para as duas amostragens e as variâncias. Quais insights você pode retirar comparando os dados?

```python
# Cálculo das médias
media_A = np.mean(amostra_estrategia_A)
media_B = np.mean(amostra_estrategia_B)

# Cálculo das variâncias
variancia_A = np.var(amostra_estrategia_A, ddof=1)  # Variância amostral (ddof=1)
variancia_B = np.var(amostra_estrategia_B, ddof=1)

print(f"Média da Estratégia A: {media_A:.2f}")
print(f"Média da Estratégia B: {media_B:.2f}")
print(f"Variância da Estratégia A: {variancia_A:.2f}")
print(f"Variância da Estratégia B: {variancia_B:.2f}")
```

- **Comparação das variâncias:**
  A variância da Estratégia A (129.27) é maior do que a da Estratégia B (110.47), sugerindo que os resultados dos alunos na Estratégia A foram mais dispersos em torno da média. Isso pode indicar que o desempenho dos alunos na Estratégia A foi mais inconsistente, enquanto a Estratégia B apresentou maior uniformidade.

**Implicações:**

- A diferença nas médias sugere que a Estratégia B pode ser mais eficaz em melhorar o desempenho médio dos alunos.
- A menor variância da Estratégia B pode indicar que ela não apenas aumenta o desempenho médio, mas também promove resultados mais consistentes entre os alunos, o que é desejável em um contexto educacional.

#### Nível de Significância (α)

O nível de significância (α) é geralmente definido pelo pesquisador antes do teste como o limite máximo aceitável para o risco de cometer um erro do tipo I (rejeitar a hipótese nula quando ela é verdadeira). No contexto deste exercício, o valor padrão usado foi α = 0.05, que é comum em estudos estatísticos.

#### Por que α = 0.05?

- **Tradição Estatística:** α = 0.05 é amplamente utilizado na pesquisa científica como um compromisso entre rigor estatístico e praticidade. Ele representa uma tolerância de 5% para o risco de rejeitar uma hipótese nula verdadeira.

- **Interpretação Prática:** Um α de 0.05 significa que estamos dispostos a aceitar até 5% de chance de que a diferença observada seja devida ao acaso, quando na realidade não há diferença significativa entre as estratégias.

- **Flexibilidade:** Embora 0.05 seja comum, o nível de significância pode ser ajustado dependendo do contexto. Em áreas com maior risco associado a decisões erradas (como medicina), pode-se usar um α mais rigoroso (por exemplo, 0.01). Já em estudos exploratórios ou menos críticos, um α maior (como 0.10) pode ser aceitável.
